# a) Download Iris_Data.csv File.
# b) Load and fetch data from file using data frame.

```python
import numpy as np
import pandas as pd

from google.colab import drive
drive.mount("/content/drive")
path = "/content/drive/MyDrive/Data/Iris.csv"
iris = pd.read_csv(path)
iris
```

# c) Analyse the species names and remove "Iris" word from 'species'.

```python
iris["Species"] = iris["Species"].str.replace("Iris-","")
iris
```

# d) Deter mine mean, median, mode, quantiles, ranges (max-min) for each petal and sepal measurement and the number of each species.

```python
stat_summary = iris.groupby('Species')[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].agg(['mean','median','quantile',lambda x: x.max() - x.min(), lambda x:x.mode().iloc[0]])
stat_summary
```

# e) Calculate the following for each species:

i. The mean of each feature measurements in a separate dataframe.

ii. The median of each of these measurements in a separate dataframe.

```python
mean_stat = iris.groupby('Species')[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].mean()
mean_stat
```

```python
median_stat = iris.groupby('Species')[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].median()
median_stat
```

```python
import matplotlib.pyplot as plt
# Scatter plot for "sepal length" vs "sepal width"
plt.scatter(iris['SepalLengthCm'],iris['SepalWidthCm'])
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
# Histogram of "sepal_length"
plt.hist(iris['SepalLengthCm'], bins=10)
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
plt.show()
```

```python
# Create subplots for histograms of all features
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.hist(iris['SepalLengthCm'], bins=10)
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')

plt.subplot(2, 2, 2)
plt.hist(iris['SepalWidthCm'], bins=10)
plt.xlabel('Sepal Width')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Width')

plt.subplot(2, 2, 3)
plt.hist(iris['PetalLengthCm'], bins=10)
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Length')

plt.subplot(2, 2, 4)
plt.hist(iris['PetalWidthCm'], bins=10)
plt.xlabel('Petal Width')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Width')

plt.tight_layout()
plt.show()
```

```python
# Create separate histograms for each feature
plt.figure(figsize=(12, 3))

plt.subplot(1, 4, 1)
plt.hist(iris['SepalLengthCm'], bins=10)
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')

plt.subplot(1, 4, 2)
plt.hist(iris['SepalWidthCm'], bins=10)
plt.xlabel('Sepal Width')
plt.ylabel('Frequency')

plt.subplot(1, 4, 3)
plt.hist(iris['PetalLengthCm'], bins=10)
plt.xlabel('Petal Length')
plt.ylabel('Frequency')

plt.subplot(1, 4, 4)
plt.hist(iris['PetalWidthCm'], bins=10)
plt.xlabel('Petal Width')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
# Boxplots for each measurement
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.boxplot(x='Species', y='SepalLengthCm', data=iris)
plt.title('SepalLengthCm')
plt.xlabel('Species')
plt.ylabel('Sepal Length')

plt.subplot(2, 2, 2)
plt.title('SepalWidthCm')
sns.boxplot(x='Species', y='SepalWidthCm', data=iris)
plt.xlabel('Species')
plt.ylabel('Sepal Width')

plt.subplot(2, 2, 3)
plt.title('PetalLengthCm')
sns.boxplot(x='Species', y='PetalLengthCm', data=iris)
plt.xlabel('Species')
plt.ylabel('Petal Length')

plt.subplot(2, 2, 4)
plt.title('PetalWidthCm')
sns.boxplot(x='Species', y='PetalWidthCm', data=iris)
plt.xlabel('Species')
plt.ylabel('Petal Width')

plt.tight_layout()
plt.show()
```

```python
# Boxplot of all measurements with species colored differently
iris_df = iris
iris_df.drop(['Id'], axis=1, inplace=True)
plt.figure(figsize=(12, 6))
sns.boxplot(x='variable', y='value', hue='Species', data=pd.melt(iris_df, id_vars=['Species']))
plt.xlabel('Measurement')
plt.ylabel('Value')
plt.title('Boxplot of Measurements by Species')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(iris_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Pairplot
sns.pairplot(iris_df, hue='Species', markers=["o", "s", "D"])
plt.show()
```
